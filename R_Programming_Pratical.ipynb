**For Loop Implementation and While Loop Implementation**

**Experiment:**

- Fibonacci series using while loop (user defined function)

- Multiplication table using for loop

- Simple calculator in R

####🔹 Program – 1
> Fibonacci Series using While Loop (User Defined Function)

```python
%%R

# User defined function for Fibonacci series
fibonacci <- function(n) {
  a <- 0
  b <- 1
  i <- 1

  cat("Fibonacci Series:\n")

  while (i <= n) {
    cat(a, " ")
    c <- a + b
    a <- b
    b <- c
    i <- i + 1
  }
}

# Function call
fibonacci(10)
```

#### 🔹 Program – 2
> Multiplication Table using For Loop

## **📘 Practical–9**

**Scoping Rules in R**

- Demonstrate local and global variables using functions

- Define a variable outside a function and modify it inside the function

- Display values before and after function execution

> Program

```python
%%R

# Global variable
x <- 10

# Display value before function call
cat("Value of x before function execution:", x, "\n")

# Function definition
change_value <- function() {
  # Local variable (same name)
  x <- 20
  cat("Value of x inside function (local):", x, "\n")
}

# Function call
change_value()

# Display value after function call
cat("Value of x after function execution:", x, "\n")
```

## **📘 Practical–10**

**Apply Family Functions**

- Use apply(), lapply(), sapply(), and tapply()

- **Experiment:** Use apply(), lapply(), and sapply() on a matrix and compare the results

```python
%%R

# Create a matrix
mat <- matrix(1:9, nrow = 3)

print("Matrix:")
print(mat)

# apply() - sum of rows and columns
row_sum <- apply(mat, 1, sum)
col_sum <- apply(mat, 2, sum)

print("Row-wise Sum using apply():")
print(row_sum)

print("Column-wise Sum using apply():")
print(col_sum)

# lapply() - apply sum to each column
lapply_result <- lapply(as.data.frame(mat), sum)
print("Result using lapply():")
print(lapply_result)

# sapply() - simplified version of lapply
sapply_result <- sapply(as.data.frame(mat), sum)
print("Result using sapply():")
print(sapply_result)
```

## **📘 Practical–16**

**Handling Missing Data**

- Use is.na(), na.omit(), and complete.cases()

- **Experiment:** Identify and replace missing values with the mean

```python
%%R

# Create vector with missing values
v <- c(10, 20, NA, 30, NA, 40)

print("Original Vector:")
print(v)

# Identify missing values
print("Identify NA values:")
print(is.na(v))

# Calculate mean (excluding NA)
mean_value <- mean(v, na.rm = TRUE)
print(paste("Mean value:", mean_value))

# Replace NA with mean
v[is.na(v)] <- mean_value
print("Vector after replacing NA with mean:")
print(v)

# Using na.omit()
clean_vector <- na.omit(v)
print("Vector after na.omit():")
print(clean_vector)

# Using complete.cases()
complete <- complete.cases(v)
print("Complete cases:")
print(complete)
```

## **📘 Practical–20**
**Reference Classes and User-Defined Functions**

- Implement reference classes

- Perform basic arithmetic operations

- **Experiment:** Create a reference class to perform arithmetic operations

> Program
- Reference Class for Arithmetic Operations

```python
%%R

# Create Reference Class
Calculator <- setRefClass(
  "Calculator",
  fields = list(
    a = "numeric",
    b = "numeric"
  ),
  methods = list(
    add = function() {
      a + b
    },
    subtract = function() {
      a - b
    },
    multiply = function() {
      a * b
    },
    divide = function() {
      if (b != 0)
        a / b
      else
        "Division by zero not possible"
    }
  )
)

# Create object of reference class
calc <- Calculator$new(a = 20, b = 10)

# Perform operations
cat("Addition:", calc$add(), "\n")
cat("Subtraction:", calc$subtract(), "\n")
cat("Multiplication:", calc$multiply(), "\n")
cat("Division:", calc$divide(), "\n")
```

## **📘 Practical–23**

**One-Way ANOVA Using R**

- Perform and interpret one-way ANOVA

- **Experiment:** Perform one-way ANOVA on a dataset and interpret the result

Screenshot 2026-02-03 134557.png

```python
%%R

# Create dataset
marks <- c(85, 88, 90,   # Group A
           78, 82, 80,   # Group B
           92, 95, 94)   # Group C

group <- factor(c("A","A","A",
                  "B","B","B",
                  "C","C","C"))

data <- data.frame(group, marks)

print(data)

# Perform One-Way ANOVA
anova_result <- aov(marks ~ group, data = data)

# Display result
summary(anova_result)
```

```python
%%R

boxplot(marks ~ group,
        col=c("red","blue","green"),
        main="One-Way ANOVA",
        xlab="Group",
        ylab="Marks")
```

## **📘 Practical–24**

**Linear Regression Analysis**

- Implement linear regression

- Interpret model output

- Visualize results

- **Experiment:** Fit a linear regression model and plot the regression line

> Program
-  Step 1: Create Dataset

```python
%%R

study_hours <- c(2,4,6,8,10)
marks <- c(40,50,65,75,90)

data <- data.frame(study_hours, marks)

print(data)
```

> ✅ Step 2: Fit Linear Regression Model

```python
%%R

model <- lm(marks ~ study_hours, data=data)

summary(model)
```

```python
%%R

plot(study_hours, marks,
     main="Linear Regression: Study Hours vs Marks",
     xlab="Study Hours",
     ylab="Marks",
     pch=16,
     col="blue")

abline(model, col="red", lwd=2)
```
