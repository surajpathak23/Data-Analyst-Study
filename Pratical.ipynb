#### 1️⃣ Create Vector

```python
%%R

mat <- matrix(c(1,2,3,4,5,6), nrow = 2, ncol = 3)

print(mat)
```

#### 3️⃣ Create List

```python
%%R

my_list <- list(name = "Suraj", age = 21, marks = c(80, 85, 90))

print(my_list)
```

#### 4️⃣ Create Data Frame

```python
%%R

student_data <- data.frame(
        Name = c("Suraj", "Ashish", "Kamlesh", "Ravi"),
        Age = c(20, 21, 19, 22),
        Marks = c(75, 80, 85, 90)
)

print(student_data)
```

#### 5️⃣ Display Structure

```python
%%R

str(student_data)
```

#### 6️⃣ Statistical Summary

```python
%%R

summary(student_data)
```

#### 7️⃣ Nature (Type) of Data

```python
%%R

sapply(student_data, class)
```

## **📘 Practical–18**

**Indexing and Accessing Object Values**

- Demonstrate indexing systems for R objects

- Experiment:

  **1.** Extract sub-matrix whose rows have column value > 3

  **2.** Convert a matrix to a 1-dimensional array

#### 1️⃣ Create  Matrix

```python
%%R

mat <- matrix(c(1,2,3,4,5,6,7,8,9), nrow = 3, byrow = TRUE)

# Display matrix
print(mat)
```

#### 🔹 **Experiment 1**
- Extract sub-matrix whose column value > 3

```python
%%R

# Select rows where 2nd column > 3
result <- mat[mat[,2] > 3, ]

# Display result
print(result)
```

#### 🔹 **Experiment 2**
- Convert matrix to 1D array

```python
%%R

# Convert matrix to 1D array
new <- as.vector(mat)

# Display result
print(new)
```
