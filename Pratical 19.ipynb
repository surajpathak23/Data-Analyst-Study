#### **Introduction to S3 and S4 Classes**

- Create and use simple S3 and S4 objects.
##### **Experiment:** Demonstrate indexing on vectors, matrices, and data frames

#### **Create S3 object**

```python
%%R

student <- list(name = "Suraj", age = 21)
class(student) <- "student_class"

print("S3 Object:")
print(student)

# Access values
print("Access Name:")
print(student$name)
```

#### **Create S4 object**

```python
%%R

# Define S4 class
setClass("Student",
         slots = list(
           name = "character",
           age = "numeric"
         ))

# Create S4 object
s4_student <- new("Student", name = "Rahul", age = 22)

print("S4 Object:")
print(s4_student)

# Access values
print("Access Name:")
s4_student@name
```

#### **Indexing on Vector**

```python
%%R

vec <- c(10, 20, 30, 40, 50)

print("Vector:")
print(vec)

# Access 3rd element
print("3rd Element:")
print(vec[3])
```

#### **Indexing on Matrix**

```python
%%R


mat <- matrix(1:9, nrow = 3)

print("Matrix:")
print(mat)

# Access element (row 2, column 2)
print("Element (3,2):")
print(mat[3,2])
```

#### **Indexing on Data Frame**

```python
%%R


df <- data.frame(
  Name = c("Aman", "Rahul", "Priya"),
  Marks = c(70, 80, 90)
)

print("Data Frame:")
print(df)

# Access first row
print("First Row:")
print(df[1,])

# Access Marks column
print("Marks Column:")
print(df$Marks)
```
