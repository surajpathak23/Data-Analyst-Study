### 🔹 Step 1: Data Collection

```python
# check missing values
df.isnull().sum()
```

### 🔹 Step 3: Data Processing
- Check DataType

```python
# Check DataType
df.dtypes
```

### 🔹 Step 4: Data Analysis

```python
df.describe()
```

```python
df.groupby('location_rating')['rent_price'].mean()
```

### 🔹 STEP 5: Data Visualization
- 📊 Bar Chart — Bedrooms vs Average Rent

```python
import matplotlib.pyplot as plt

bedroom_rent = df.groupby('bedrooms')['rent_price'].mean()

plt.bar(bedroom_rent.index, bedroom_rent.values)
plt.xlabel("Bedrooms")
plt.ylabel("Average Rent")
plt.title("Bedrooms vs Average Rent")
plt.show()
```

- 📈 Scatter Plot — Area vs Rent

```python
plt.scatter(df['area_sqft'], df['rent_price'])
plt.xlabel("Area (sqft)")
plt.ylabel("Rent Price")
plt.title("Area vs Rent Price")
plt.show()
```

- Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

# correlation matrix
corr = df.corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of House Rent Dataset")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df.drop('rent_price', axis=1)
y = df['rent_price']

model = RandomForestRegressor()
model.fit(X, y)

importance = pd.Series(model.feature_importances_, index=X.columns)
importance.sort_values().plot(kind='barh')
plt.title("Feature Importance")
plt.show()
```

### ✅ STEP 6: Insights & Observations (Short)
- Rent price increases as the house area increases.
- Houses with more bedrooms have higher rent.
- Properties with better location rating have higher rent.
- Rent decreases as the distance from city center increases.
- Furnished houses generally have higher rent than unfurnished houses.

### ✅ STEP 7: Decision Making (Short)
- People can choose houses based on their budget and requirements.
- Houses near the city center should be selected only if the budget is high.
- For lower rent, houses farther from the city center can be preferred.
- Furnished houses can be chosen for better comfort if budget allows.
- The model can help in predicting rent price for new houses.

## **Pratical - 7**
- Create an area chart for time-series data such as yearly profit or rainfall data.

```python
df = pd.read_csv('global_sports_footwear_sales_2018_2026.csv')
df.head(5)
```

```python
# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Extract year
df['Year'] = df['order_date'].dt.year

# Calculate yearly profit (total revenue)
yearly_profit = df.groupby('Year')['revenue_usd'].sum()

# Create area chart
plt.figure(figsize=(8,5))
plt.fill_between(yearly_profit.index, yearly_profit.values)

plt.xlabel("Year")
plt.ylabel("Total Revenue (USD)")
plt.title("Yearly Profit Area Chart")

plt.show()
```

## **Pratical - 11**
- Draw a simple node-link diagram to show relationships.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(["Course", "BCA","MCA", "BBA", "B.Tech", "MBA"])

# Add relationships
G.add_edge("Course", "BCA")
G.add_edge("Course", "MCA")
G.add_edge("Course", "BBA")
G.add_edge("Course", "B.Tech")
G.add_edge("Course", "MBA")

# Draw node-link diagram
plt.figure(figsize=(7,5))
nx.draw(G, with_labels=True, node_size=3000)

plt.title("Node-Link Diagram: Department and Courses")
plt.show()
```

## **Pratical - 15**
- Create a multivariate chart using more than two variables.

```python
import pandas as pd
import matplotlib.pyplot as plt

# read dataset
df = pd.read_csv("student_performance.csv")

plt.figure(figsize=(9,6))

scatter = plt.scatter(
    df["Study_Hours"],
    df["Marks"],
    s=df["Attendance"]*8,       # bubble size based on attendance
    c=df["Sleep_Hours"],        # color based on sleep hours
    cmap="viridis",
    edgecolors="black",
    alpha=0.8
)

plt.title("Multivariate Visualization: Study Hours vs Marks", fontsize=14)
plt.xlabel("Study Hours", fontsize=12)
plt.ylabel("Marks", fontsize=12)

plt.grid(True)

# color bar
cbar = plt.colorbar(scatter)
cbar.set_label("Sleep Hours")

plt.show()
```
